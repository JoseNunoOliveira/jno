## Who were Abbot António da Costa's friends in Porto?

</span>

![image.jpg](./media/acosta.jpg)

---

**Title**:  Who were the friends of Abbot António da Costa in Porto?

**Abstract**: Thirteen letters by Abbot António da Costa (1714-1780?) have survived that span over 30 years (1750-1780). These letters were written to five friends of him in Porto and refer to no less than 97 people altogether. Of these, 21 are mentioned more than once and seem to be close friends living in Porto. Who were these close friends? This talk reports on recent research that analyses the graph (network) of personal relationships inherent in the letters and dives into its central clusters. By inspection of parish records and other documentation available from the archives of the University of Coimbra, Porto municipality and Torre do Tombo it is possible to reconstruct some of the families addressed in the letters. The main conclusion is that such families were of relatively high rank but there is still much to be known concerning the musical environment in Porto which gave birth to the violinist and composer António da Costa.

Cartas curiosas do Abbade Antonio da Costa 93 (pdf 98) 125

Nas peças que chegaram até nós, a dificuldade de compreensão e a não habitualidade das tendências musicais surgem logo à primeira leitura. A par destas, deve notar-se a estranheza dos conjuntos e o seu inusitado. Os dois O vos omnes são respectivamente para cinco vozes de soprano solo a cappella em Fá menor, e para quatro sopranos acompanhados por tantos violinos em Dó menor, o outro. Será talvez uma tentativa do autor, uma prova de perícia destinada a demonstrar a possibilidade de criar variedade através da exploração de um único registo, abrangendo-o em toda a sua extensão, evitando também a monotonia eventualmente criada por um único timbre? Ou, ao contrário, produzir um resultado alienante, de máxima homogeneidade, para o qual o efeito auditivo de monodia aparente gera polifonia?
Um trecho da peça para vozes solistas pode dar conta disso (Ex. 1). Cromatismos, modulações inesperadas e imprevisíveis – sobretudo concentradas entre os BBs. 9-13 – calúnias ininterruptas tornam a fala sempre suspensa e tensa. A escrita não é fácil para as vozes obrigadas a saltar constantemente entre intervalos incômodos ou difíceis - por serem excessivos ou diminuídos - e também há falsas relações entre as partes. Não menos audaciosa, embora em tonalidade menos obsoleta que a anterior, é a versão com acompanhamento de violino. Deve-se notar a anotação colocada no topo da partitura, «In camera», que talvez pudesse aludir ao destino da peça, à sua execução não necessariamente dentro da liturgia.
A função dos instrumentos com sua longa introdução de 27 compassos é colocar o ouvinte em um clima flutuante e, portanto, sustentar as vozes dobrando-as (Ex. 2, bb. 1-16).
A esta parte inicial, repleta de acentos melódicos sempre otimistas e que ultrapassam constantemente o âmbito do compasso, segue-se a entrada em imitação dos quatro sopranos (Ex. 3, compassos 28-39).
E a peça prossegue nesta atmosfera 'futurista' de contínua tensão e suspensão, criada pelos atrasos e apogiaturas, que também põe à prova a capacidade rítmica dos intérpretes.
Como confirmação da concepção complexa e altamente cerebral de Costa, deve ser mencionado o dueto para dois violinos,56 também numa estranha tonalidade para a época – Mi bemol menor – em que se ouve um fluxo quase imparável de figurações rítmicas e saltos de linhas melódicas apenas aqui e ali repentinamente interrompidos por pausas, apenas para dar breve descanso aos instrumentistas e ouvintes, quase atordoados pela incessante sucessão de sons e pelo entrelaçamento rítmico dos dois instrumentos mentes. O dueto é composto por três andamentos: um Allegro mas não muito que do compasso 2/4 inicial passa para o compasso 3/4 e depois volta para o compasso 2/4 mantendo sempre o mesmo andamento; um Lento ma poco bipartido e um Presto que reproduz as feições e caráter do primeiro. Em Ex. 4 (Allegro, mas não muito, compassos 1-30) alguns trechos significativos. Note-se em particular o fraseado que provoca incessantemente um descompasso entre ritmo e métrica, os sincopados prolongados e a variedade de articulações, por vezes claramente diferenciadas entre primeiro e segundo violino.
Após esta série de desenhos, a b. 45 iniciam novas figurações que continuam até b. 60 (Ex. 5) e desaguam numa outra tendência que se torna cada vez mais densa e 'vanguardista' nas harmonias e ousadias com cromatismos e modulações contínuas (Ex. 6).
Mesmo o tempo lento é dominado por síncopes e cromatismos, bem como por figurações que se intensificam ritmicamente no final de cada uma das duas partes que compõem a peça (Ex. 7, Lento ma poco, compassos 1-17).
Pode ser provável que Bu Burney refere-se precisamente a este dueto ao narrar a tentativa de execução feita pelo próprio abade e pelo Sr. Stratzel «excelente executor e bom músico» na conclusão da reunião de convívio na casa de Lord Stormont. «O dueto do Abade Costa era tão difícil tanto em termos de ritmo como de estilo, que mesmo depois de vinte ou trinta ensaios não conseguiam uma boa execução».57 A conclusão do episódio é evidente: música impraticável até para o próprio compositor, de natureza puramente especulativa, concebida para uma reflexão exclusivamente teórica e para um puro exercício contrapontístico.

Por outro lado, no que diz respeito aos traços peculiares do seu estilo, observa-se sobretudo a homogeneidade do conjunto (apenas violinos, apenas sopranos, com exceção única do moteto para quatro vozes ímpares) que contrasta com a heterogeneidade do figurações, dos metros/ritmos, articulações, harmonias desdobradas ao longo da página. Além disso, a escolha de uma tonalidade estranha ao lado da fixidez tonal dos diferentes andamentos que compõem uma peça não é secundária; as modulações conduzidas segundo uma lógica que satisfaz o ouvido (atingindo-o também com fortes dissonâncias) e não respeita os princípios rigorosos do sistema tonal que se estabeleceu a nível teórico desde há algumas décadas. Com efeito, como relata Burney, Costa declarou-se inimigo da teoria de Rameau, acreditando «que o baixo fundamental é a mais absurda de todas as invenções porque destrói toda fantasia, todo nexo e toda continuidade, tendendo continuamente para uma cadência final, isto é, a resolução de qualquer sentença iniciada».64 E para «um homem talvez mais extravagante na natureza do que na música»65 isto era verdadeiramente insuportável. No entanto, a sua lição deixa um traço não tão desprezível no catálogo de Tartini que, em homenagem e imitação «acima do estilo que toca a prette da chittara portuguesa», escreve a sonata em Lá maior Lá 4.66 dominada talvez mais que outras, por movimentos móveis e sobretudo por passagens imitativas da guitarra (particularmente em estilo rasgueado) presentes no terceiro dos quatro andamentos que o compõem (Allegro, Andante, Allegro, Minuetto). A datação desta página, atualmente presumida entre 1730-1750, face à dedicatória, presente porém apenas na cópia conservada em Berkeley (italiano MS 773), deve ser adiada para a década de 1960, precisamente o período em que o dois músicos eram relacionamento próximo. Igualmente concebível seria o caso de apenas a cópia de Berkeley datar desse período e isso testemunharia que Tartini havia encontrado em da Costa um interlocutor em harmonia com os princípios harmônicos que vinha cultivando em seu trabalho teórico desde os anos 1950 e que esse sonata mais do que outras, embora composta em época anterior à presença dos dois, poderia ser a demonstração prática mais evidente, bem como uma homenagem ao personagem.
